```python
data_gold=data["Gold"].value_counts()
```

```python
dataset_corr=data.copy()
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
for column in dataset_corr.columns:
    if dataset_corr[column].dtype == 'object':
        dataset_corr[column] = label_encoder.fit_transform(dataset_corr[column])

correlation_matrix=dataset_corr.corr()
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
